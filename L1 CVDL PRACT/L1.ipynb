```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img_bgr = cv2.imread('lufffy.jpg')
```

```python
img = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img[:,:,1]
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(gray)
```

```python
fig = plt.figure(figsize=(8, 6))

plt.subplot(1,4,1)
plt.title("Original rgb")
plt.imshow(img)
plt.axis('off')

plt.subplot(1,4,2)
plt.title("gray 0")
plt.imshow(img[:,:,0],cmap='gray')
plt.axis('off')

plt.subplot(1,4,3)
plt.title("gray 1")
plt.imshow(img[:,:,1],cmap='gray')
plt.axis('off')

plt.subplot(1,4,4)
plt.title("gray 2")
plt.imshow(img[:,:,2],cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

#### Image Modification

```python
resized = cv2.resize(img, (300, 300))
```

```python
resized.shape
```

```python
plt.imshow(resized)
```

```python
cv2.line(resized,(0,0),(150,150),(255,0,0),2)
cv2.rectangle(resized, (150, 150), (200, 200), (0, 255, 0), 3)
cv2.circle(resized, (225, 225),35, (0, 0, 255), -1)
plt.imshow(resized)
plt.show()
```
